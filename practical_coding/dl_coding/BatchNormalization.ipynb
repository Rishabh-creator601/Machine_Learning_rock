```python
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_circles
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Conv2D,MaxPool2D,Flatten,BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping

import numpy as np 
import matplotlib.pyplot as plt
```

```python
x,y =  make_circles(n_samples=10000,noise=0.2)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
y_train.shape
```

```python
model = Sequential([
    Dense(64,input_dim=2,activation="relu"),
    Dense(32,input_dim=2,activation="relu"),
    Dense(1,activation="sigmoid")
])
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=['acc'])
history = model.fit(x_train,y_train,epochs=100,validation_data=(x_test,y_test))
```

```python
model_batch = Sequential([
    Dense(64,input_dim=2,activation="relu"),
    BatchNormalization(),
    Dense(32,input_dim=2,activation="relu"),
    BatchNormalization(),
    Dense(1,activation="sigmoid")
])
model_batch.compile(loss="binary_crossentropy",optimizer="adam",metrics=['acc'])
history_batch = model_batch.fit(x_train,y_train,epochs=100,validation_data=(x_test,y_test))
```

```python
plt.figure(figsize=(12,8))
plt.style.use("ggplot")
plt.subplot(1,2,1)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(["train", "val"],loc='upper left')
plt.title("Normal Model")

plt.subplot(1,2,2)
plt.plot(history_batch.history['loss'])
plt.plot(history_batch.history['val_loss'])
plt.legend(["train", "val"],loc='upper left')
plt.title("BatchNormalization Model")
```
