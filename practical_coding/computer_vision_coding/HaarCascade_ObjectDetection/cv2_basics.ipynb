# Loading a image


**cv2.IMREAD_GRAYSCALE** :  0 <br />
**cv2.IMREAD_COLOR** :  1  <br />
**cv2.IMREAD_UNCHANGED** :  -1 <br/>

cv2.IMREAD_UNCHANGED Reads alpha channel : In short which offers transparency 

**what is alpha channel ?** <br/> 
The alpha channel is a special channel that handles transparency. When an image has an alpha channel on it, it means you can adjust the image's opacity levels and make bits translucent or totally see-through. The alpha channel is instrumental when you want to remove the background from an image.


```python
image = cv2.imread("rot.jpg",1)
print(image.shape)
print(image)
```

The below image displayed is different from original image **why ?** <br />
**Opencv** stores image in **BGR format** whereas **Matplotlib** show image in **RGB format** <br />
Lets Correct it 

```python
plt.imshow(image)
```

**Corrected Image**

```python
image = image[:,:,::-1]
plt.imshow(image)
```

#### Lets explore color channels

```python
b,g,r = cv2.split(image)


plt.figure(figsize=[20,5])
plt.subplot(161);plt.imshow(r);plt.title("Red Channel");
plt.subplot(162);plt.imshow(g);plt.title("Green Channel");
plt.subplot(163);plt.imshow(b);plt.title("Blue Channel");


merged = cv2.merge((r,g,b))
plt.subplot(164);plt.imshow(merged);plt.title("RGB Channel");

merged2 = cv2.merge((b,g,r))
plt.subplot(165);plt.imshow(merged2);plt.title("BGR Channel");
```

#### Converting BGR to RGB via **cvtColor**

```python
img  =  cv2.cvtColor(merged2,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

### Exploring HSV Channels <br />

H - HUE    <br />
S - SATURATION   <br />
V - VALUE

```python
hsvImage= cv2.cvtColor(image,cv2.COLOR_RGB2HSV)

h,s,v = cv2.split(hsvImage)

plt.figure(figsize=[20,5])
plt.subplot(161);plt.imshow(h);plt.title("HUE Channel");
plt.subplot(162);plt.imshow(s);plt.title("SATURATION Channel");
plt.subplot(163);plt.imshow(v);plt.title("VALUE Channel");

s = s + 70

plt.subplot(164);plt.imshow(s);plt.title("SATURATION new Channel");
```
