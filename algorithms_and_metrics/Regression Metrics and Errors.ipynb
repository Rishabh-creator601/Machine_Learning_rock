```python
import numpy as np 
import pandas as pd 
import seaborn  as sns
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 
from help import find_file 
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error,r2_score
```

```python
data  = pd.read_csv(find_file("placement.csv"))
```

```python
x_train,x_test,y_train,y_test = train_test_split(data['cgpa'],data['package'])
```

```python
x_train =  np.array(x_train).reshape(-1,1)
x_test =  np.array(x_test).reshape(-1,1)
y_train  =  np.array(y_train).reshape(-1,1)
```

```python
## Video source :  https://www.youtube.com/watch?v=Ti7c-Hz7GSM&list=PLKnIA16_Rmvbr7zKYQuBfsVkjoLcJgxHH&index=52 
## Adjusted r2 score and r2 score is imp 
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_predict =  model.predict(x_test)
```

```python
print("MAE : ",mean_absolute_error(y_test,y_predict))
print("MSE : ",mean_squared_error(y_test,y_predict))
print("RMSE  : ",root_mean_squared_error(y_test,y_predict))
print("R2 SCORE   : ",r2_score(y_test,y_predict))
```

```python
## adjusted r2 score 
n  = x_test.shape[0]
k =  x_test.shape[1] -  1

adjusted_r2   =  1 -  ((1-r2_score(y_test,y_predict))*(n -1 )/(n - 1 - k))
```
