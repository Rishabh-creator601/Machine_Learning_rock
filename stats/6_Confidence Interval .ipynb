```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

## **MARGIN OF ERROR** = $Z_{\alpha/2}$ $\frac{\sigma}{\sqrt{n}}$
=>  where n >= 30  <br />
**We will se the following relations**
### 1.  ME $\propto \sigma$
### 2.  ME $\propto 1/\sqrt{n}$
 ### 3. ME $\propto Z_{\alpha/2}$ 

```python
### 1st One 

confidence_level =  0.95  # FIXED
sample_size =  30  # FIXED 
pop_std_devs =  np.arange(1,101) ##  Make variations 

#z_score =  (1-0.95)/2 +0.95 = > (1+0.95)/2 
critical_value = stats.norm.ppf((1+confidence_level)/2 )
margin_of_errors =  critical_value * (pop_std_devs/np.sqrt(sample_size))


plt.plot(pop_std_devs,margin_of_error)
plt.xlabel("Std of Population  ")
plt.ylabel("Margin of ERROR")
plt.grid(True)
```

```python
### 2nd One 

confidence_level =  0.95  # FIXED
sample_size = np.arange(30,101) ## MAKE VARIATIONS  
pop_std_devs = 20   # FIXED

#z_score =  (1-0.95)/2 +0.95 = > (1+0.95)/2 
critical_value = stats.norm.ppf((1+confidence_level)/2 )
margin_of_errors =  critical_value * (pop_std_devs/np.sqrt(sample_size))


plt.plot(sample_size,margin_of_errors)
plt.xlabel("Sample Sizes ")
plt.ylabel("Margin of ERROR")
plt.grid(True)
```

```python
### 3rd  One 

confidence_levels =  np.arange(0.50,1.1,0.01).round(1)
sample_size =  30  # FIXED
pop_std_devs = 20   # FIXED

#z_score =  (1-0.95)/2 +0.95 = > (1+0.95)/2 
critical_value = stats.norm.ppf( (1 +confidence_levels )/2  )
margin_of_errors =  critical_value * (pop_std_devs/np.sqrt(sample_size))


plt.plot(confidence_levels,margin_of_errors)
plt.xlabel("Confidence Levels  ")
plt.ylabel("Margin of ERROR")
plt.grid(True)
```
