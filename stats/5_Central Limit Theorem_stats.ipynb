```python
import numpy as np 
import random 
import matplotlib.pyplot as plt
```

**CENTRAL LIMIT THEOREM**

-  convert any Dist to Normal Dist
-  if the dist has $\mu$ and $\sigma^2$ then New Normal Dist has $\mu$ and $\sigma^2/n$
-  `NOTE : we are talking about` $\sigma^2$ `Not` $\sigma$ 

### Converting Uniform -> Normal

```python
low =  0
high =  2
size = (1000,300) # bigger the size[1] the more the dist move towards NORMAL DIST
samples =  np.random.uniform(low,high,size)

samples_means = np.mean(samples,axis=1)

plt.hist(samples_means,bins=30,density=True )
```

```python
print(f"MEAN BEFORE NORMAL DIST: {samples.mean()}")
print(f"MEAN AFTER NORMAL DIST : {samples_means.mean()}")
```

```python
print(f"VAR BEFORE NORMAL DIST: {samples.var()}")
print(f"VAR AFTER NORMAL DIST : {samples_means.var()*size[1]}")
```

### Converting Exponential -> Normal 

```python
size = (1000,300)
scale =  2
exp_Dist =  np.random.exponential(scale=1/scale,size=size)

dist_means = np.mean(exp_Dist,axis=1)
plt.hist(dist_means,density=True,bins=30)
```

```python
print(f"MEAN BEFORE NORMAL DIST: {exp_Dist.mean()}")
print(f"MEAN AFTER NORMAL DIST : {dist_means.mean()}")
```

```python
print(f"VAR BEFORE NORMAL DIST: {exp_Dist.var()}")
print(f"VAR AFTER NORMAL DIST : {dist_means.var()*size[1]}")
```
