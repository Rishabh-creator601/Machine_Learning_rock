```python
import pandas as pd
import matplotlib.pyplot  as plt 
import seaborn as sns 
import os 
```

```python
data  =  os.path.join("C:\\codes\\data\\")
tips  =  pd.read_csv(data  + "tip.csv")
titanic =  pd.read_csv(data + "titanic\\train.csv")
iris = pd.read_csv(data  + "iris.csv")
```

```python
tips.head()
```

```python
titanic =  titanic.drop(columns=['Name','Ticket','Cabin'])
titanic.head()
```

```python
iris.head()
```

### Scatterplot  (Numerical - Numerical)

```python
# One 
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
# two
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
#++
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='smoker')
```

```python
#++
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='smoker',size='size')
```

### BarPlot (Categorical - Numerical)

```python
sns.barplot(data=titanic,x='Pclass',y='Age')
```

```python
sns.barplot(data=titanic,x='Pclass',y='Age',hue='Sex')
```

### BoxPlot  (Categorical - Numerical)

```python
sns.boxplot(data=titanic,x='Sex',y='Age',hue='Survived')
```

### Distplot 

```python
##Very detail full graph 
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

### HeatMap

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
## This shows people in P class 1 survived more whereas in P class they died more 
```

### Cluster Map 

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

### PairPlot 

```python
sns.pairplot(data=iris,hue='Species')
```
